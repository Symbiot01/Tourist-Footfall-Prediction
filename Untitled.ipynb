```python
# (Optional) If pytrends is needed to fetch Google Trends data:
# !pip install pytrends
from pytrends.request import TrendReq

# Function to fetch Google Trends for a list of terms
def fetch_trends(terms, start_date, end_date, geo=''):
    """
    Fetch Google Trends interest over time for given search terms between start_date and end_date.
    Returns a DataFrame with each term's interest (0-100 scaled) indexed by date.
    """
    pytrends = TrendReq(hl='en-US', tz=360)
    timeframe = f"{start_date} {end_date}"
    pytrends.build_payload(terms, timeframe=timeframe, geo=geo)
    data = pytrends.interest_over_time()
    # Drop the 'isPartial' column if present
    if 'isPartial' in data.columns:
        data = data.drop('isPartial', axis=1)
    return data
```

```python
!pip install pytrends

import os
import time
import pandas as pd
from datetime import datetime, timedelta
from pytrends.request import TrendReq
from dateutil.relativedelta import relativedelta

# ------------------------- CONFIGURATION ------------------------- #

# Categories and associated search terms (max 5 per request)
grouped_terms = {
    'Accommodation': [
        'hotels in Shimla', 'budget hotels in Shimla', 'resorts near Shimla',
        'luxury hotels in Shimla', 'homestays in Shimla', 'best places to stay in Shimla'
    ],
    'Transportation': [
        'how to reach Shimla', 'Delhi to Shimla by road', 'Chandigarh to Shimla train',
        'Shimla toy train booking', 'Shimla bus timetable', 'Shimla airport connectivity'
    ],
    'Tour Planning': [
        'Shimla travel guide', 'Shimla itinerary 3 days', 'Shimla trip cost',
        'Shimla tour package', 'Shimla honeymoon package', 'solo travel to Shimla'
    ],
    'Attractions': [
        'places to visit in Shimla', 'Mall Road Shimla', 'Jakhoo temple Shimla',
        'Kufri near Shimla', 'Christ Church Shimla', 'Shimla sightseeing'
    ],
    'Activities': [
        'things to do in Shimla', 'adventure activities in Shimla', 'shopping in Shimla',
        'skiing near Shimla', 'nature walks in Shimla', 'trekking in Shimla'
    ],
    'Food & Dining': [
        'street food in Shimla', 'best restaurants in Shimla', 'Shimla local food',
        'Shimla food tour', 'cafes in Shimla'
    ],
    'Seasonal & Weather': [
        'Shimla in summer', 'Shimla in winter', 'best time to visit Shimla',
        'Shimla snowfall forecast', 'Shimla weather today'
    ],
    'Events & Culture': [
        'Shimla summer festival', 'Shimla Christmas celebration', 'Shimla cultural events',
        'history of Shimla', 'British architecture in Shimla'
    ]
}

# Time Range
start_date = datetime.strptime("01-01-2010", "%d-%m-%Y")
end_date = datetime.strptime("31-12-2023", "%d-%m-%Y")

# Output Path
output_path = "summed_trends_by_category.csv"

# ------------------------- HELPERS ------------------------- #

def month_range(start, end):
    current = start
    while current <= end:
        yield current
        current += relativedelta(months=1)

def fetch_trends(terms, start, end, geo='IN'):
    pytrends = TrendReq(hl='en-US', tz=360)
    timeframe = f"{start} {end}"
    pytrends.build_payload(terms, timeframe=timeframe, geo=geo)
    df = pytrends.interest_over_time()
    return df.drop(columns=['isPartial'], errors='ignore')

# ------------------------- MAIN LOGIC ------------------------- #

monthly_data = pd.DataFrame()

for category, terms in grouped_terms.items():
    print(f"\nFetching category: {category}")
    category_monthly = pd.Series(dtype='float64')

    for i in range(0, len(terms), 5):  # Max 5 terms/request
        subset = terms[i:i+5]
        print(f"  Subset: {subset}")
        subset_monthly = pd.Series(dtype='float64')

        for month_start in month_range(start_date, end_date):
            month_end = (month_start + relativedelta(months=1)) - timedelta(days=1)
            try:
                time.sleep(2)  # Rate limit guard
                df = fetch_trends(subset, month_start.strftime('%Y-%m-%d'), month_end.strftime('%Y-%m-%d'))
                if df.empty:
                    continue
                daily_sum = df.sum(axis=1)  # Sum across terms
                month_total = daily_sum.sum()
                subset_monthly.at[month_start.strftime('%d-%m-%Y')] = month_total
            except Exception as e:
                print(f"     Error at {month_start.strftime('%Y-%m')} for {subset}: {e}")

        category_monthly = category_monthly.add(subset_monthly, fill_value=0)

    if monthly_data.empty:
        monthly_data = category_monthly.to_frame(name=category)
    else:
        monthly_data[category] = category_monthly

# Final formatting
monthly_data.reset_index(inplace=True)
monthly_data.rename(columns={'index': 'Date'}, inplace=True)
monthly_data = monthly_data[['Date'] + [col for col in monthly_data.columns if col != 'Date']]

# Export
monthly_data.to_csv(output_path, index=False)
print(f"\n Exported full monthly data to: {output_path}")
```
